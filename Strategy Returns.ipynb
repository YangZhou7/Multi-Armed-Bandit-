```python
from data.data_process import *

from Strategies.EqualWeight import EqualWeight
from Strategies.MarketCap import MarketCap 
import matplotlib.pyplot as plt
import datetime as dt
```

```python
TRAIN_DATE_START = dt.datetime(2012, 1, 1)
TRAIN_DATE_END = dt.datetime(2014, 1, 1)
TRAIN_STOCK_NAMES = ['AMZN',
                     'AAPL',
                     'MSFT',
                     'AMD']
data_source = 'yahoo'
data_folder = './data/' + data_source
data = get_data(data_folder + '/test_data.npy', type='test', source=data_source,date_start=TRAIN_DATE_START,
                      date_end=TRAIN_DATE_END,stock_names=TRAIN_STOCK_NAMES)
```

```python
equal_weight=EqualWeight(data)
market_cap=MarketCap(data)
```

```python
plt.plot(equal_weight.sum(axis=1))
plt.plot(market_cap.sum(axis=1))
plt.legend(['Equal Weight','Market Cap'])
plt.show()
```
